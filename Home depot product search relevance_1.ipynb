```python
from sentence_transformers import SentenceTransformer

model = SentenceTransformer('all-mpnet-base-v2')
```

```python
search_term_embed = model.encode(df['search_term'].to_numpy())
print(search_term_embed)
```

```python
product_title_embed = model.encode(df['product_title'].to_numpy())
print(product_title_embed)
```

```python
product_description_embed = model.encode(df['product_description'].to_numpy())
print(product_description_embed)
```

```python
df['search_term_embed'] = search_term_embed.tolist()
df['product_title_embed'] = product_title_embed.tolist()
df['product_description_embed'] = product_description_embed.tolist()
print(df.columns)
df.head()
```

```python
import numpy as np

x = np.hstack((search_term_embed, product_title_embed, product_description_embed))
# x = np.hstack((search_term_embed, product_description_embed, result))
# x = df['cosine_sim'].to_numpy()
print(x.shape)
print(x)
y = df['median_relevance'].to_numpy()
print(y.shape)
print(y)
```

```python
from keras.utils.np_utils import to_categorical   
from sklearn.preprocessing import LabelEncoder

# # y = [1,4,3,2]
# y = y.astype(np.float)
label_encoder = LabelEncoder()
y_new = label_encoder.fit_transform(y)

print(type(y_new[0]))
print(y_new)
y_new = to_categorical(y_new, num_classes=4)
print(y_new)
```

```python
## IMPLEMENTING QUADRATIC WEIGHTED KAPPA IN NEURAL NETWORK

from tensorflow import keras 
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
import math
# Hyperparameters
num_epochs=25
batch_size=32
input_dim=len(x[0])
print(input_dim)
neurons = math.floor((2*len(x))/2)
print(neurons)
output_dim = 3

# # Define custom loss

def sklearn_qwk(y_true, y_pred) -> np.float64:
  tf.config.run_functions_eagerly(True)
  y_true_new = y_true.numpy()
  y_pred_new = y_pred.numpy()
  # print(type(y_true_new))
  y_true_new = label_encoder.inverse_transform(argmax(y_true, axis = 1))
  y_pred_new = label_encoder.inverse_transform(argmax(y_pred, axis = 1))

  return cohen_kappa_score(y_true_new, y_pred_new, weights="quadratic")

# End of quadratic weighted kappa metric

# model_cohen_kappa = cohen_kappa_loss(num_classes=4)
import sklearn.metrics as metrics

#define the model
model = Sequential()
model.add(Dense(neurons, input_dim = input_dim, activation = 'relu'))
model.add(Dense(output_dim, activation = 'softmax'))


model.compile(loss = 'categorical_crossentropy', optimizer = 'adam', metrics = ['accuracy', sklearn_qwk])

model.summary()
```
